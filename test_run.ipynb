# 1. Standard Bar Chart

Not recommended to use if there are only 2 labels

```python
labels = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8','Q9', 'Q10', 'Q11', 'Q12']
values = [15, 23, 12, 34, 22, 18, 27, 30,23,40,15,27]

plots.NUHS_bar_chart(labels, values, x_label = 'x', y_label = 'y', title = 'Gradient Bar Chart')
```

# 2. Stacked Bar Chart

```python
labels = ["Q1", "Q2", "Q3", "Q4"]
category_names = ["Product A", "Product B", "Product C", "Product D"]
data = [
    [10, 15, 20, 25],  # Product A
    [5,  10, 15, 10],  # Product B
    [2,  5,  10, 5],    # Product C
    [3,  16,  5, 9]    # Product C
]

plots.NUHS_stacked_bar_chart(labels, data, category_names)
```

# 3. Grouped Bar Chart

Two categories will only display blue and orange colour 

```python
labels = ['Q1', 'Q2', 'Q3', 'Q4']
category_names = ['Product A', 'Product B']
data = [
    [20, 35, 30, 35],
    [25, 32, 34, 20]
]

plots.NUHS_grouped_bar_chart(labels, data, category_names)
```

```python
labels = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5']
category_names = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
data = [
    [20, 35, 30, 35, 40],
    [25, 32, 34, 20, 30],
    [15, 22, 18, 25, 28],
    [10, 20, 15, 18, 22]
]

plots.NUHS_grouped_bar_chart(labels, data, category_names)
```

# 4. Scatter plot

```python
x_groups = [
    [18.5, 22.3, 24.8],    # Group 1
    [27.2, 30.5, 33.1, 35.6]  # Group 2
]

y_groups = [
    [110, 115, 120],
    [126, 133, 140, 150], 
]

category_names = ["Normal BMI", "High BMI"]
# labels_groups = [["P1", "P2", "P3"], ["P4", "P5", "P6", "P7"]]

plots.NUHS_scatter_plot(
    x_groups,
    y_groups,
    category_names,
    x_label = 'x',
    y_label = 'y',
    title="BMI vs SBP by Category"
)
```

# 5. Line Plot

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
cardiology = [120, 135, 150, 160, 155, 170]
oncology = [80, 85, 90, 95, 100, 110]
pediatrics = [100, 105, 95, 90, 100, 105]

y_list = [cardiology, oncology, pediatrics]
category_names = ['Cardiology', 'Oncology', 'Pediatrics']

plots.NUHS_line_plot(months, y_list, category_names, title="Monthly Patient Count by Department")
```
